```python
import pandas as pd
import matplotlib.pyplot as plt

future_predictions_df = pd.read_csv('future_predictions.csv')

top_20_predictions = future_predictions_df.sort_values(by='predicted_price_in_5_years', ascending=False).head(20)

plt.figure(figsize=(12, 8))

plt.barh(top_20_predictions['Make'] + ' ' + top_20_predictions['Model'], top_20_predictions['predicted_price_in_5_years'], color='skyblue')

plt.xlabel('Predicted Price in 5 Years')
plt.ylabel('Car Make and Model')
plt.title('Top 20 Predicted Car Prices in 5 Years (Descending Order)')

plt.gca().invert_yaxis()

plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 8))

plt.barh(top_20_predictions['Make'] + ' ' + top_20_predictions['Model'], top_20_predictions['predicted_price_in_5_years'], color='skyblue')

plt.xlabel('Predicted Price in 5 Years (Log Scale)')
plt.ylabel('Car Make and Model')
plt.title('Top 20 Predicted Car Prices in 5 Years (Descending Order)')

plt.xscale('log')

plt.gca().invert_yaxis()

# Show the plot
plt.show()
```
